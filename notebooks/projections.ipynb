The purpose of this notebook is to explore how projections work. We will start by projecting a cube into 2-D space

```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt

# Generate the cube
cube_x = np.linspace(-1, 1, 10)
cube_y = np.linspace(-1, 1, 10)
cube_z = np.linspace(-1, 1, 10)
cube_x, cube_y, cube_z = np.meshgrid(cube_x, cube_y, cube_z)

# Look at one projection
plt.imshow(cube_x[:, :, 0])
```

```python
# View using interactive 3D projection in Jupyter

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(cube_x, cube_y, cube_z, c=cube_z, marker='o')
```

```python
cube = np.mgrid[0:5, 0:5, 0:5].reshape(3,-1).T

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(cube[:,0], cube[:,1], cube[:,2], marker='o')
```

```python
theta = np.pi / 4
proj_matrix = np.array(
    [
        [1, 0],
        [0, 1],
        [0, 0],
    ]
)
rotation_matrix = np.array(
    [
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)],
    ]
)
cube_2d_proj = cube @ proj_matrix @ rotation_matrix
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(cube_2d_proj[:,0], cube_2d_proj[:,1], s=4)
```
